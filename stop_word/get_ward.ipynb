```python
import urllib3

slothlib_path = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/' \
                'Filter/StopWord/word/Japanese.txt'
http = urllib3.PoolManager()
r = http.request('GET', slothlib_path)
slothlib_file = r.data.decode('utf-8')
slothlib_stopwords = [word for word in slothlib_file.split('\r\n') if word != '']
print(slothlib_stopwords)
```
